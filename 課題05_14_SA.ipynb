# **プログラムの名称**
ガチャプログラム
## **概要**

私の作ったオリジナルガチャを引くことができるプログラムです。このプログラムでは自分の引き当てた武器や衣装を、自分自身に装備することができ、装備するとHPや攻撃力が上がります。

## **操作方法**
まず、フォームに衣装のガチャ武器のガチャをそれぞれ何回引くかを半角数字で入力してください。その後、プログラムを実行するとガチャの結果が表示されます。獲得できたアイテムが一覧で表示されるので、あなたが装備したいアイテムを一つ選んで入力してください。エンターキーを押すと残りのプログラムが実行され、あなたの最終的なステータス（後述）が表示されます。

```python
#@title サブ処理

import random

# ユーザーに衣装ガチャの回数を入力させる
#@markdown **■ 衣装ガチャを引く回数**（半角数字）
outfit_draw_count = 2 #@param{type:"number"}

# ユーザーに武器ガチャの回数を入力させる
#@markdown **■ 武器ガチャを引く回数**（半角数字）
weapon_draw_count = 3 #@param{type:"number"}


# 衣装アイテムと重みの辞書
outfit_items = ["よろい", "かぶと", "まんと", "むぎわらぼうし"]
outfit_weights = [0.4, 0.3, 0.2, 0.1]
outfit_dict = dict(zip(outfit_items, outfit_weights))

# 衣装ごとのHP増加量の辞書
outfit_hp_increase = {"よろい": 100, "かぶと": 50, "まんと": 20, "むぎわらぼうし": 10}

# 武器アイテムと重みの辞書
weapon_items = ["スナイパー", "ショットガン", "ケン", "オノ"]
weapon_weights = [0.4, 0.3, 0.2, 0.1]
weapon_dict = dict(zip(weapon_items, weapon_weights))

# 武器ごとの攻撃力増加量の辞書
weapon_attack_increase = {"スナイパー": 150, "ショットガン": 100, "ケン": 50, "オノ": 10}

# ガチャ結果の個数をカウントするリスト（初期値はすべて0）
outfit_counts = [0] * len(outfit_items)
weapon_counts = [0] * len(weapon_items)

# 所持金
money = 3000

# HP
hp = 3000

# 攻撃力
attack_power = 100

# ガチャを引く関数
def play_gacha(gacha_type, draw_count):
    global money, outfit_counts, weapon_counts, hp, attack_power

    if gacha_type == "衣装":
        gacha_dict = outfit_dict
        items = outfit_items
        counts = outfit_counts
        price = 100
    elif gacha_type == "武器":
        gacha_dict = weapon_dict
        items = weapon_items
        counts = weapon_counts
        price = 500
    else:
        print("無効なガチャの種類です。")
        return

    # ガチャを引く回数が所持金より大きい場合、ガチャを終了
    if price * draw_count > money:
        print("所持金が不足しています。ガチャを終了します。")
        return

    results = []  # ガチャの結果を一時的に保存するリスト

    for _ in range(draw_count):
        money -= price
        item = random.choices(items, weights=list(gacha_dict.values()))[0]
        item_index = items.index(item)
        counts[item_index] += 1
        results.append(item)

    if gacha_type == "衣装":
        print("--------------------------------------------------------")
        print("衣装ガチャの結果：")
        for outfit in results:
            print(outfit)

        print("どの衣装を着ますか？")
        available_outfits = [outfit for outfit, count in zip(outfit_items, outfit_counts) if count > 0]
        for outfit in available_outfits:
            print(outfit)
        chosen_outfit = input("着る衣装を選んでください：")
        while chosen_outfit not in available_outfits:
            print("選択が無効です。")
            chosen_outfit = input("着る衣装を選んでください：")

        hp_increase = outfit_hp_increase[chosen_outfit]
        hp += hp_increase
        print(f"あなたは{chosen_outfit}を選びました。")
        print(f"HPが{hp_increase}増加しました。")

    elif gacha_type == "武器":
        print("--------------------------------------------------------")
        print("武器ガチャの結果：")
        for weapon in results:
            print(weapon)

        print("どの武器を装備しますか？")
        available_weapons = [weapon for weapon, count in zip(weapon_items, weapon_counts) if count > 0]
        for weapon in available_weapons:
            print(weapon)
        chosen_weapon = input("装備する武器を選んでください：")
        while chosen_weapon not in available_weapons:
            print("選択が無効です。")
            chosen_weapon = input("装備する武器を選んでください：")

        attack_increase = weapon_attack_increase[chosen_weapon]
        attack_power += attack_increase
        print(f"あなたは{chosen_weapon}を装備しました。")
        print(f"攻撃力が{attack_increase}上昇しました。")
```

```python
#@title メイン処理

# 衣装ガチャを引く
play_gacha("衣装", outfit_draw_count)



# 武器ガチャを引く
play_gacha("武器", weapon_draw_count)

# アイテムの個数を表示
print("--------------------------------------------------------")
print("現在の衣装の所持数：")
for item, count in zip(outfit_items, outfit_counts):
    print(item, ":", count)

print("--------------------------------------------------------")
print("現在の武器の所持数：")
for item, count in zip(weapon_items, weapon_counts):
    print(item, ":", count)

# ステータスの表示
print("--------------------------------------------------------")
print("あなたのステータス：")
print("あなたの現在のHP：", hp)
print("あなたの現在の攻撃力：", attack_power)
print("現在の残高：", money)
```
